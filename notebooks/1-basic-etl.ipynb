# Series

```python
# initialize data dictionary

dict_data = {
    "IDMF": 5,
    "TROF": 2,
    "GODF": 1,
    "DETN": 1
}
```

```python
# create data series
series_data = pd.Series(dict_data)

# print data
series_data
```

```python
# initialize data using list

list_data = [5, 2, 1, 1]
index_data = ["IDMF", "TROF", "GODF", "DETN"]
```

```python
# convert the data list into a data series
series_data_2 = pd.Series(list_data, index=index_data)

# show data series
series_data_2
```

# Dataframe

```python
# create data in the form of a Python data dictionary
dict_data = {
    "item_name": ["Indomie", "Taro", "Good Day", "Detergent"],
    "item_type": ["food", "food", "food", "non-food"],
    "quantity": [5, 2, 1, 1]
}

# create a dataframe from the data dictionary
df = pd.DataFrame(dict_data)

# display data
df
```

```python
# create data in the form of a Python data dictionary
dict_data = {
    "item_name": ["Indomie", "Taro", "Good Day", "Detergent"],
    "item_type": ["food", "food", "food", "non-food"],
    "quantity": [5, 2, 1, 1]
}

# initialize index values
index_data = ["IDMF", "TROF", "GODF", "DETN"]

# create a dataframe from the data dictionary and add the index parameter
df2 = pd.DataFrame(dict_data, index=index_data)

# display data
df2
```

# API / JSON

```python
import requests
```

```python
# get API data from PokeAPI
resp = requests.get("https://pokeapi.co/api/v2/pokemon?limit=500")

# check the status code of the API, 200 means it returned a response
resp.status_code
```

```python
resp.text
```

```python
# convert API data to JSON format
raw_data = resp.json()

# display the raw data
raw_data
```

```python
pokemon_data = pd.DataFrame(raw_data['results'])

pokemon_data
```
